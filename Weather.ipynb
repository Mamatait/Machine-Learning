```python

import numpy as np
import pandas as pd
```

```python
Weather_celcius=pd.Series([36,38,24,21,39,40,np.nan,-3,-4,11])
print(Weather_celcius)
Weather_celcius.isnull() #Weather with 0 degree/celcius
Weather_celcius.notnull() #Weather more than 0 celcius
```

```python
Weather_data={
    #1:Europe 2:North America 3:Antartica 4:Asia 5:Australia
    '1 Europe':[36,40,-1,np.nan,21],
    '2 America':[np.nan,-3,-4,np.nan,11],
    '3 Antartica':[np.nan,np.nan,np.nan,np.nan,np.nan],
    '4 Asia':[40,36,24,np.nan,18],
    '5 Australia':[49,39,np.nan,np.nan,10]
    }
df=pd.DataFrame(Weather_data)
df
```

```python
#Fill all Nan as 0
df_fill = df.fillna(0)
df_fill
```

```python
#Want to fill the  temperature data which I got
df_fill = df.fillna({'1 Europe': 10, '2 America': 60, '3 Antartica': -10, '4 Asia': 40, '5 Australia':30})
df_fill
```
